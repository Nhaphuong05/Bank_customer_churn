```python
import pandas as pd
import numpy as np

#1. import file
file_path = "C:/Users/hello/Downloads/capstone (1)/capstone/data/Bank-Customer-churn.csv"
df = pd.read_csv(file_path)

print(f"Dữ liệu ban đầu: {df.shape[0]} dòng")
```

```python
#2. Làm sạch dữ liệu
#Loại bỏ duplicates dựa trên CustomerId
df = df.drop_duplicates(subset='CustomerId')
# Xử lý missing values (nếu có)
# Kiểm tra xem có cột nào bị thiếu dữ liệu không
print("Số lượng giá trị thiếu mỗi cột:\n", df.isnull().sum())
# Thường các cột số dư (Balance) nếu trống ta điền 0, còn lại xóa các dòng trống thông tin quan trọng
df['Balance'] = df['Balance'].fillna(0)
df = df.dropna()
# Xử lý Outliers (Giá trị ngoại lai) cho Age và Balance
# Age: Giới hạn trong khoảng 18 - 90 tuổi
# Balance: Giới hạn tối thiểu là 0 (không để số dư âm)
df['Age'] = df['Age'].clip(18, 90)
df['Balance'] = df['Balance'].clip(lower=0)
```

```python
# 3. TẠO CỘT MỚI (FEATURE ENGINEERING) - Đề bài yêu cầu
# Tạo cột Age Group: Young (<30), Middle-Aged (30-45), Senior (45-60), Elderly (>60)
age_bins = [0, 30, 45, 60, 100]
age_labels = ['Young', 'Middle-Aged', 'Senior', 'Elderly']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)

# Tạo cột Balance Category: Low (<50k), Medium (50k-100k), High (>100k)
balance_bins = [-1, 50000, 100000, float('inf')]
balance_labels = ['Low', 'Medium', 'High']
df['Balance Category'] = pd.cut(df['Balance'], bins=balance_bins, labels=balance_labels)
```

```python
# 4. XUẤT FILE SẠCH
# File này dùng để nạp vào Power BI thực hiện Task 2 và 3
df.to_csv('Bank_Churn_Cleaned.csv', index=False)

print("--- XỬ LÝ HOÀN TẤT ---")
print(df[['CustomerId', 'Age', 'Age Group', 'Balance', 'Balance Category']].head())
```
